This is a continuation of "Capturing Images with Webcam". We will be using almost exactly the same code, except we'll put our read() function within a controlled loop.

```python
#import libriaries
import cv2 as cv

# initialize the camera
cap = cv.VideoCapture(0)
#check that it worked properly, otherwise exit
if not cap.isOpened():
    print("Cannot open camera")
    exit()

#loop to capture our video
while True:
    # Capture frame-by-frame
    ret, frame = cap.read()
    # check that frame was read correctly
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break

    # Display the resulting frame
    cv.imshow('frame', frame)
    
    #print fps
    framespersecond= int(cap.get(cv.CAP_PROP_FPS))
    print("The total number of frames in this video is ", framespersecond)
    
    
    #If 'q' is pressed then exit loop
    if cv.waitKey(1) == ord('q'):
        break

# When everything done, release the capture
cap.release()
cv.destroyAllWindows()
```

We can also access some features of the video such as Width, Height, and Frame rate using the get(propoid) function. For the full list of Video Capture Properties go to : https://docs.opencv.org/4.x/d4/d15/group__videoio__flags__base.html#gaeb8dd9c89c10a5c63c139bf7c4f5704d
